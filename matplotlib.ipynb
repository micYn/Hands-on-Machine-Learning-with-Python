```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y=x+1
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,11)
print(x)
y = np.geomspace(0.1,1,num=10)
z = np.logspace(0.1,1,num=10,base=10)
print(y,z)
plt.plot(x,y); plt.plot(x,z)     # plt.plot(x, y, x, z) -> same
plt.show()
```

```python
t = np.arange(0, np.pi*2, 0.05)
print(t)
y = np.sin(3*t)
plt.plot(t,y)
plt.show()
```

```python
n = 5
x = np.arange(n)
y = np.random.rand(n)   # rand: uniform distribution. specify dimension in para
plt.bar(x,y)
plt.show()
```

```python
# plt.subplots() demo
fig, axs = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 cols
fig.suptitle('The big title')

axs[0].plot([1, 2, 3], [1, 4, 9])
axs[0].set_title("First plot")
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

axs[1].plot([1, 2, 3], [9, 4, 1])
axs[1].set_title("Second plot")

plt.tight_layout()
plt.show()
```

```python
x = np.arange(10)
# plt.subplots() demo2
fig, axs = plt.subplots(2,2)
axs[0][0].plot(x,x)
axs[0][0].set_title('Linear')

axs[0][1].plot(x,x*x)
axs[0][1].set_title('Quadratic')

axs[1][0].plot(x,np.sqrt(x))
axs[1][0].set_title('Square root')

axs[1][1].plot(x,np.log(x))
axs[1][1].set_title('Log')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```

```python
# Same result of previous cell in a MATLAB style
# Unlike plt.subplots(), which returns both the figure and axes objects for multiple subplots, 
# plt.subplot() creates and activates a specific subplot within a grid.
x = np.arange(10)

plt.subplot(2,2,1) # 2 rows, 2 cols, index 1(first subplot of four subplots)
plt.plot(x,x); plt.title('Linear')

plt.subplot(2,2,2)
plt.plot(x,x*x); plt.title('Quadratic')

plt.subplot(2,2,3)
plt.plot(x,np.sqrt(x)); plt.title('Square root')

plt.subplot(2,2,4)
plt.plot(x,np.log(x)); plt.title('Log')

plt.tight_layout()
plt.show()
```

```python
# scatter plot
n=10
x = np.random.rand(n)
y = np.random.rand(n)
plt.scatter(x,y)
plt.show()
```

```python
# histogram (good for representing the frequency distribution of data)
mu, sigma = 0, 0.1
x = np.random.normal(mu, sigma, 1000)
print(x)
plt.hist(x)
plt.show()
```
