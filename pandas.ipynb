```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Pandas Series

```python
s1 =pd.Series([1,2,3,4,5])
print(type(s1))
s2 = pd.Series(np.arange(5), dtype=np.uint8)
print(s2)
# left column: indices
# right columns: data
```

```python
arr1 = np.array([1,3,5,7,9], dtype=np.uint8)
s3 = pd.Series(arr1, dtype=np.int16)
print(s3)
# The data type of the series will be considered as the final data type.
```

```python
# To see the values in the series:
print(s3.values)
s3.array
```

```python
# To see the index of the series
s3.index
```

```python
s3.dtype    #shape, size, nbytes, ndim
```

#### Pandas Dataframes

```python
# Like tables in databases 
# We can create dataframes from series, Ndarrays, lists, and dictionaries.
# A dictionary:
data = {'city': ['Bangalore', 'Bangalore', 'Bangalore',
                'Mumbai', 'Mumbai', 'Mumbai'],
       'year': [2020, 2021, 2022, 2020, 2021, 2022,],
       'population': [10.0, 10.1, 10.2, 5.2, 5.3, 5.5]}
df1 = pd.DataFrame(data)
print(df1)
```

```python
df1.head()
```

```python
# Specify the order of columns:
df2 = pd.DataFrame(data, columns=['year','city', 'population'])
print(df2)
```

#### Visualizing the data in DataFrames

```python
df1 = pd.DataFrame()
df1['A'] = pd.Series(list(range(100)))
df1['B'] = np.random.randn(100,1) # 100x1 matrix filled with random numbers from standard normal distribution
df1
```

```python
# Objects of Pandas data structures call Matplotlib visualization functions like plot(). 
df1.plot(x='A', y='B')
plt.show()
```

```python
df2 = pd.DataFrame(np.random.rand(10, 4),
                   columns=['A', 'B', 'C', 'D'])
print(df2[['A']])

df2.plot.bar()
df2.plot.bar(stacked=True)
df2.plot.barh()
plt.show()
```

```python
df2.plot.hist(alpha=0.7)    # alpha is the transparency of the plot
df2.plot.hist(stacked = True, alpha=0.7) # better
df2.plot.hist(stacked=True, alpha=0.7, bins=20) # the data will be divided into 20 intervals along the x-axis
                                                # Each bin represents a range of values
plt.show()
```

```python
df2.plot.box()
df2.plot.area()
df2.plot.area(stacked=False)
plt.show()
```

```python
df = pd.DataFrame({
    'Category1': [3, 2, 5, 7, 6],
    'Category2': [4, 7, 2, 4, 5],
    'Category3': [2, 4, 6, 8, 5]
}, index=[1, 2, 3, 4, 5])

# Plot an area chart
df.plot.area(alpha=0.6)

plt.title('Area Plot Example')
plt.show()
```
